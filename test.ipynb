```python
# List of 2024 MLB team abbreviations
# Will need to change to something more flexible in the future
# Team names, locations can change
mlb_teams = [
    'ATL',
    'AZ',
    'BAL',
    'BOS',
    'CHC',
    'CIN',
    'CLE',
    'COL',
    'CWS',
    'DET',
    'HOU',
    'KC',
    'LAA',
    'LAD',
    'MIA',
    'MIL',
    'MIN',
    'NYM',
    'NYY',
    'OAK',
    'PHI',
    'PIT',
    'SD',
    'SEA',
    'SF',
    'STL',
    'TB',
    'TEX',
    'TOR',
    'WSH',
]

# Hit Types
hit_types = [
    'single', 
    'double', 
    'triple',  
    'home_run',
]

# Testing with the NYY win over HOU 7-1
# Set the date range
my_team = 'NYY'
opp_team = 'HOU'
start_date = '2024-03-29'
end_date = start_date
```

```python
# Fetch Statcast data for the date range
data = statcast(start_dt=start_date, end_dt=end_date, team = opp_team)
```

```python
data.events.unique()
```

```python
data[data['events'].isin(hit_types)][['batter', 'hit_distance_sc']]
```
